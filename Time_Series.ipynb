```python
#import the libraries
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla=pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head(3)
```

```python
df_tesla.plot()
```

```python
df_tesla['High'].plot(xlim=['2020-01-01','2022-02-14'],figsize=(12,4),c ='green',ls='--')
```

```python
index=df_tesla.loc['2020-01-01':'2021-01-01'].index
share_open=df_tesla.loc['2020-01-01':'2021-01-01']['Open']
```

```python
share_open
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure,axis=plt.subplots()
axis.plot(index,share_open)
```

```python
##Time Resample
```

```python
df_tesla.resample(rule='QS').max()
```

```python
#Quartely frequencyp
df_tesla.resample(rule='QS').max().plot()
```

```python
##Business end freuency
df_tesla.resample(rule='BA').max()
```

```python
##Business Quarters freuency
df_tesla.resample(rule='BQS').max()
```

```python
df_tesla['High'].rolling(10).mean()
```

```python
df_tesla['Open: 30 days rolling']=df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head()
```

```python
df_tesla[['Open','Open: 30 days rolling']].plot(figsize=(12,5))
```
